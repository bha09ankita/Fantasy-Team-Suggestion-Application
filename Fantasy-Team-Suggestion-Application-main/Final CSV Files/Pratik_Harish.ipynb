```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
data
```

```python
data = data[['id', 'inning', 'batsman', 'bowler', 'total_runs', 'is_wicket']]
```

```python
data.head(197)
```

```python
# checking null values
data.isna().sum()
```

```python
data_dummy = data
data_dummy
```

```python
# for calculatin fantasy points, adding constraints of boundries
# for 4's adding +1 in the fantasy points and for 6's adding +2 in the fatasy points
data_dummy.loc[data_dummy['total_runs'] == 6, 'total_runs'] = data_dummy.loc[data_dummy['total_runs'] == 6, 'total_runs'] + 2
data_dummy.loc[data_dummy['total_runs'] == 4, 'total_runs'] = data_dummy.loc[data_dummy['total_runs'] == 4, 'total_runs'] + 1
```

```python
data_dummy.head(197)
```

```python
data_dummy["total_runs"].dtypes
```

```python
data_dummy
```

```python
data_dummy = data_dummy.groupby(['batsman', 'id']).sum()
```

```python
data_dummy
```

```python
# calculating fantasy points of batsman by subtracting fantasy points of wickets
data_dummy['fantasy_point_batting'] = data_dummy['total_runs'] - (data_dummy['is_wicket']*2)
```

```python
data_dummy
```

```python
# fantasy points scored by batsman from 2008-2020
data_dummy
```

```python
# calculating fantasy points for bowlers
data_dummy_2 = data[['id', 'bowler', 'inning', 'is_wicket']]
data_dummy_2 = data_dummy_2.rename(columns={'is_wicket':'wicket_taken', 'inning':'balls_bowled'})
data_dummy_2
```

```python
data_dummy_2 = data_dummy_2.groupby(['bowler', 'id']).sum()
data_dummy_2
```

```python
# as for each wicket, bowlers scores 25 fantasy points, multiplying wickets by 25 to get fantasy points of each bowler
data_dummy_2['fantasy_point_bowling'] = data_dummy_2['wicket_taken'] * 25
data_dummy_2
```

```python
data_dummy
```

```python
# merging the data of batsman and bowler
data_merged= pd.concat([data_dummy, data_dummy_2], axis=1)
data_merged
```

```python
# filling NaN values by 0 
data_merged = data_merged.fillna(0)
data_merged
```

```python
# adding fantasy points of batsman and bowler to get the final fantasy points
data_merged['total_fantasy_points'] = data_merged['fantasy_point_batting'] + data_merged['fantasy_point_bowling']
```

```python
data_merged
```

```python
data_merged = data_merged.rename(columns={'inning':'balls_played'})
# data_merged = data_merged.drop(columns=['fantasy_point_batting', 'fantasy_point_bowling'])
data_merged
```

```python
# saving final fantasy points into csv file
data_merged.to_csv("fantasy_point_2008_2020.csv")
```

```python
df = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
df
```

```python
df = df[['inning', 'batsman', 'bowler', 'batsman_runs', 'total_runs', 'is_wicket']]
df
```

```python
df_batsman = df[['inning', 'batsman', 'batsman_runs', 'is_wicket']]
df_batsman
```

```python
df_batsman = df_batsman.groupby('batsman').sum()
df_batsman = df_batsman.rename(columns={'inning':'balls_played'})
df_batsman
```

```python
df_batsman['batting_avg'] =df_batsman['batsman_runs'] // df_batsman['is_wicket']
df_batsman
```

```python
df_batsman['batting_avg'] = df_batsman['batting_avg'].replace([np.inf, -np.inf], 0)
```

```python
df_batsman
```

```python
df_batsman['batting_strike_rate'] = df_batsman['batsman_runs'] * 100 // df_batsman['balls_played']
df_batsman
```

```python
df_batsman = df_batsman.drop(columns=['batsman_runs'])
df_batsman
```

```python
df_bowler = df[['inning', 'bowler', 'total_runs', 'is_wicket']]
df_bowler = df_bowler.rename(columns={'inning':'balls_bowled'})
df_bowler
```

```python
df_bowler = df_bowler.groupby('bowler').sum()
df_bowler
```

```python
df_bowler['bowling_avg'] = df_bowler['total_runs'] // df_bowler['is_wicket']
df_bowler['bowling_avg'] = df_bowler['bowling_avg'].replace([np.inf, -np.inf], 0)
df_bowler
```

```python
df_bowler['bowling_strike_rate'] = df_bowler['balls_bowled'] // df_bowler['is_wicket']
df_bowler = df_bowler.replace([np.inf, -np.inf], 0)
df_bowler
```

```python
df_bowler = df_bowler.drop(columns=['balls_bowled', 'total_runs'])
df_bowler
```

```python
df_batsman
```

```python
df_players_merged = pd.concat([df_batsman, df_bowler], axis=1)
df_players_merged = df_players_merged.fillna(0)
df_players_merged
```

```python
df_players_merged.to_csv("players_previous_record.csv")
```

```python
df_pre_rec = pd.read_csv('players_previous_record.csv')
df_pre_rec
```

```python
df_calc_score = df_pre_rec
```

```python
df_calc_score = df_calc_score.rename(columns={'batting_avg':'calc_runs'})
df_calc_score = df_calc_score.drop(columns=['balls_played', 'wickets_taken'])
df_calc_score
```

```python
df_calc_score['calc_balls_played'] = df_calc_score['calc_runs'] * df_calc_score['batting_strike_rate'] // 100
df_calc_score
```

```python
df_calc_score['calc_wickets_taken'] = df_calc_score['bowling_avg'] / 24
df_calc_score
```

```python
df_calc_score['calc_balls_bowled'] = df_calc_score['bowling_strike_rate'] * df_calc_score['calc_wickets_taken']
df_calc_score['calc_wickets_taken'] = df_calc_score['calc_wickets_taken'].round(0)
df_calc_score['calc_balls_bowled'] = df_calc_score['calc_balls_bowled'].round(0)
df_calc_score
```

```python
df_calc_score = df_calc_score[['players', 'calc_balls_played', 'calc_runs', 'is_wicket', 'calc_balls_bowled', 'calc_wickets_taken']]
```
