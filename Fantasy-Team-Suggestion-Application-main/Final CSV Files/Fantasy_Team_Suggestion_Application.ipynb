### Creating a function to give six players from both the teams who can perform better
#### Here, we are creating a function which takes input as the players of the both the teams, we are giving them preference according to our batsman dataset mentioned above. Then we are sorting this dataset according to the preference number given

```python
# creting an empty dataframe
fantasy_batsman = pd.DataFrame()
fantasy_batsman
```

```python
# we are adding player name and team name along with the fantasy_points to a dataframe
# this will be our dataset which contains the sorted players of one team
for i in team_csk.index:
    temp = team_csk['batsman'][i]
    for j in data.index:
        if temp == data['players'][j]:
            fantasy_batsman = fantasy_batsman.append({'fantasy_points':data['fantasy_points'][j], 'name':temp, 'team':team_csk['batting_team'][i]}, ignore_index=True)
```

```python
# sorting the players by fantasy points in descending order and removing duplicates if there any
fantasy_batsman = fantasy_batsman.sort_values('fantasy_points')
fantasy_batsman = fantasy_batsman.drop_duplicates()
fantasy_batsman
```

```python
# now, we are adding next team's player to our dataset and then we are sorting the dataset on the basis of reference id
# and we are taking only first six players
for i in team_mi.index:
    temp = team_mi['batsman'][i]
    for j in data.index:
        if temp == data['players'][j]:
            fantasy_batsman = fantasy_batsman.append({'fantasy_points':data['fantasy_points'][j], 'name':temp, 'team':team_mi['batting_team'][i]}, ignore_index=True)
```

```python
# sorting the players by fantasy points in descending order and removing duplicates if there any
fantasy_batsman = fantasy_batsman.drop_duplicates()
fantasy_batsman
fantasy_batsman = fantasy_batsman.sort_values('fantasy_points', ascending=False)
```
