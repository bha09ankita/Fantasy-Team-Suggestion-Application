```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
data
```

```python
data.batsman.nunique()
```

```python
df_batsman = data.groupby('batsman').sum()
df_batsman
```

```python
df_batsman = df_batsman.drop(['id','extra_runs','total_runs','non_boundary','ball','over'],axis =1)
df_batsman
```

```python
df_batsman['strike'] = data.groupby('batsman').mean()['batsman_runs']*100
```

```python
df_batsman
```

```python
df_batsman['batting_avg'] = df_batsman['batsman_runs']/df_batsman['is_wicket']
df_batsman
```

```python
df_batsman['batting_avg'] = df_batsman['batting_avg'].replace([np.inf, -np.inf], 0)
df_batsman
```

```python
df_batsman
```

```python
#df_batsman = df_batsman.sort_values(by =['batsman_runs','batting_avg','strike'],ascending=False)
df_batsman = df_batsman.sort_values(by =['batting_avg','batsman_runs','strike'],ascending=False)
df_batsman.to_csv('final_batsman_dataset.csv')
```
