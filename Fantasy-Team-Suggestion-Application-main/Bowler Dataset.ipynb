```python
import pandas as pd
```

```python
import numpy as num
```

```python
data = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
data
```

```python
data = data[['inning', 'over', 'ball', 'bowler', 'total_runs', 'is_wicket']]
data
```

```python
df_bowler = data.groupby('bowler').sum()
df_bowler
```

```python
df_bowler['overs_bowl'] = df_bowler['inning']/6
df_bowler
```

```python
df_bowler['bowling_avg'] = df_bowler['total_runs']/df_bowler['is_wicket']
df_bowler
```

```python
df_bowler['bowling_strike_rate'] = df_bowler['inning']/df_bowler['is_wicket']
df_bowler
```

```python
bowler_database_trial = df_bowler.sort_values(by=['bowling_avg', 'total_runs'], ascending=True)
bowler_database_trial
```

```python
bowler_dataset = bowler_database_trial.sort_values(by=['is_wicket'], ascending=False)
bowler_dataset
```

```python
bowler_dataset.to_csv('bowler_dataset.csv')
```
